```python
import numpy as np
import matplotlib.pyplot as plt

# Define the given function
def given_function(x):
    return (np.log(x) / (np.sin(2 * x) + 1.5)) + x / 5

# Define the range of x
x_values = np.linspace(2, 10, 100)  # 100 points between 2 and 10
y_values = given_function(x_values)

# Plotting the function
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='f(x) = (ln(x) / (sin(2x) + 1.5)) + x/5')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of the Given Function')
plt.legend()
plt.grid(True)
plt.show()
```
