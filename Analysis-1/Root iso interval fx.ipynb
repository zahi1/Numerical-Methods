```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# given function
def fx(x):
  return -0.63*x**4 + 3.92*x**3 - 7.95*x**2 + 5.50*x - 0.53
```

```python
# initial data
xmin, xmax = -1, 3.05
h = 0.3

# ploting givven function
dx = 0.05;  x=np.arange(xmin,xmax, dx); y=fx(x)
plt.xlabel("x"); plt.ylabel("y"); plt.plot(x, y, 'b'); plt.grid()

# bracketing roots
while xmin<xmax:
  if np.sign(fx(xmin)) !=np.sign(fx(xmin+h)):
    print("Root : [%.3f ;  %.3f]" % (xmin,xmin+h))
    plt.plot([xmin], [0], 'or')
    plt.plot([xmin+h], [0 ], 'og')
  xmin += h
```
