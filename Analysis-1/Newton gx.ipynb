```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# given function
def fx(x):
  return np.sin(x) * (x**2 - 1) * (x + 3) - 0.9
```

```python
#data to visulize function in interval [3,4]
xmin, xmax = -2,5 # points for visualization
# ploting givven function
dx = 0.05;  x=np.arange(xmin,xmax, dx); y=fx(x)
plt.xlabel("x"); plt.ylabel("y"); plt.plot(x, y, 'b'); plt.grid()
# ploting bracketing points
xs = 3.5
plt.plot([xs], [0], 'or')
```

Newton

```python
def dfx(x):
  return np.cos(x)*(x**2 - 1)*(x + 3) + 2*x*np.sin(x)*(x + 3) + np.sin(x)*(x**2 - 1)
```

```python
eps = 1e-8
xi = xs
while np.abs(fx(xi)) > eps:
  xi_bef = xi
  xi = xi - (1 / dfx(xi)) * fx(xi)
  print("Fx = " + str(fx(xi)) + "  /  x = " + str(xi))
  plt.xlabel("x"); plt.ylabel("y"); plt.plot(x, y, 'b'); plt.grid()
  plt.plot([xi_bef], fx(xi_bef), 'ob')
  plt.plot([xi], [0], 'or')
  plt.plot([xi_bef, xi], [fx(xi_bef), 0], 'r-')
  plt.plot([xi, xi], [0, fx(xi)], 'g--')
  plt.show()
print("done")
```
