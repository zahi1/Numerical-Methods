```python
import matplotlib.pyplot as plt
from shapely.geometry import Polygon
```

```python
geom = []
filename =  "kaunas.txt"
with open(filename) as f:
  str = [word for line in f for word in line.split(", ")]
geom = []
for i in range(len(str)):
  lon, lat = str[i].split(" ")
  geom.append((float(lon), float(lat)))
p1 = Polygon(geom)
plt.plot(*p1.exterior.xy)
```

```python
pointsType = {'P1': 'r', 'P2': 'g', 'P3': 'b', 'P4': 'c', 'A1': 'm', 'A2': 'y', 'A3': 'k', 'A4': 'w' }
filename =  "part_of_data_from_appendix.txt" # example only from one region
currentLocations = {}
file1 = open(filename, 'r')
lines = file1.readlines()
nr = 0
for line in lines:
    locationType, y, x = line.split()
    currentLocations[nr] = {'locationType':locationType, 'y' : float(y), 'x': float(x)}
    nr+=1

# ploting 
p1 = Polygon(geom)
plt.plot(*p1.exterior.xy)
for currentLocationNr in currentLocations:
    plt.plot([currentLocations[currentLocationNr]['x']], [currentLocations[currentLocationNr]['y']], 'o'+pointsType[currentLocations[currentLocationNr]['locationType']])
```
