```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits import mplot3d
import math
```

```python
def LF(x):  # returns values vector
    s = np.array([2 * np.sin(x[0]) + x[0] + x[1], 
                  4 * np.cos(2 * x[1]) - x[1] + 0.5 * x[0]])
    return s
```

```python
#surfaces

xx=np.linspace(-5,5,200);
yy=np.linspace(-5,5,200);

Z=np.zeros(shape=(len(xx),len(yy),2))
X, Y = np.meshgrid(xx, yy);
for i in range (0,len(xx)):
  for j in range (0,len(yy)): 
      Z[i,j,:]=LF([X[i][j],Y[i][j]]).transpose();


ax = plt.axes(projection ='3d') 
surf1 = ax.plot_surface(X, Y, Z[:,:,0], color='blue', alpha=0.4, linewidth=0.1, antialiased=True)
CS11 =  ax.contour(X, Y, Z[:,:,0],[0],colors='b')
plt.show()

ax = plt.axes(projection ='3d') 
surf1 = ax.plot_surface(X, Y, Z[:,:,1], color='red', alpha=0.4, linewidth=0.1, antialiased=True)
CS11 =  ax.contour(X, Y, Z[:,:,1],[0],colors='r')
plt.show()
```
